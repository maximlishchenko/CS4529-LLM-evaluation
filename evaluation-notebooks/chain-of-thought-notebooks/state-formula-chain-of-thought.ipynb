# Task 3.7a

Prompt type: chain-of-thought

Task: for a trace, state the formula that was used to get to the final emission score. Use labels as entity names and 
conversion factor IDs as their values.

Given as example: smli_trace1_valid

Given chain-of-thought process: The formula can be extracted by inspecting emission calculation activities in the trace, which are represented 
as "https://w3id.org/peco#EmissionCalculationActivity".
To get the formula for the emission score, we need to find the calculation activity the emission score was 
generated by. In the given example trace, this is the activity with id "https://github.com/mlco2/impact/provenance/i/2932985a-f353-43e8-b3ec-e85d1ad9ed58"
This activity has used two entities in the calculation, namely:
"https://github.com/mlco2/impact/provenance/i/CalculationEntity/9a6a5950-b38e-489b-b599-644d99e79965" and
"https://w3id.org/ecfkg/i/mlco2/azure/centralindia/cf".
"https://w3id.org/ecfkg/i/mlco2/azure/centralindia/cf" is a conversion factor value which was not generated by 
anything, so we keep track of this entity. "https://github.com/mlco2/impact/provenance/i/CalculationEntity/9a6a5950-b38e-489b-b599-644d99e79965" 
was generated by another activity with id "https://github.com/mlco2/impact/provenance/i/bec867c8-e5a1-4014-9408-a83fc67ce28b"
Next, we inspect this activity. It used the following entities in its calculation:
"https://github.com/mlco2/impact/provenance/i/CalculationEntity/ff8ac588-a1ea-4aa0-9f50-b77a4ef2995d" and 
"https://github.com/mlco2/impact/provenance/i/CalculationEntity/ObservationResult/7514bae6-3232-4aeb-b1ca-f073949063a0".
"https://github.com/mlco2/impact/provenance/i/CalculationEntity/ff8ac588-a1ea-4aa0-9f50-b77a4ef2995d" was not 
generated by anything and has a label "Watt Consumption", we keep track of that.
"https://github.com/mlco2/impact/provenance/i/CalculationEntity/ObservationResult/7514bae6-3232-4aeb-b1ca-f073949063a0" 
was also not generated by any activity and has a label "Duration of Use". Hence, the activity above performed the 
calculation (Watt Consumption * Duration of Use). By inspecting the associated values we can see that the 
result of this calculation was also divided by 1000. So the resulting formula for 
"https://github.com/mlco2/impact/provenance/i/bec867c8-e5a1-4014-9408-a83fc67ce28b" is ((Watt Consumption * Duration of Use) / 1000).
Lastly, we append to this formula the conversion factor that we recorded above.
Hence, the result is ((Watt Consumption * Duration of Use) / 1000) * https://w3id.org/ecfkg/i/mlco2/azure/centralindia/cf

Given as task: smli_trace3_def

Expect to see in response:

One of the following:
1. ((Watt Consumption * Duration of use) / 1000) * "https://w3id.org/ecfkg/i/mlco2/aws/cn-north-1/cf"
2. ((Watt Consumption * Duration of use) / 1000) * 0.68

Quantified as: 1
